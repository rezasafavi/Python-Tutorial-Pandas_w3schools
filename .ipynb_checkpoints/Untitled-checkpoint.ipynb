```python
import pandas as pd
mydataset = {
  'cars': ["BMW", "Volvo", "Ford"],
  'passings': [3, 7, 2]
}

myvar = pd.DataFrame(mydataset)

print(myvar)
```

```python
# Create a simple Pandas Series from a list:
a = [1,2,7]
mayvar = pd.Series(a)
print(a)
```

```python
# Return the first value of the Series:
a = [1,2,7]
myvar = pd.Series(a)
print(myvar[0])
```

```python
# Create you own labels:
a = [1, 7, 2]

myvar = pd.Series(a, index = ["x", "y", "z"])
print(myvar)
```

```python
# Return the value of "y":
print(myvar["y"])
```

```python
# Create a simple Pandas DataFrame:
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)

print(df)
```

```python
# Return row 0:
print(df.loc[0])
```

```python
# Return row 0 and 1:
print(df.loc[[0, 1]])
```

```python
# Add a list of names to give each row a name:
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df = pd.DataFrame(data, index = ["day1", "day2", "day3"])

print(df) 
```

```python
# Return "day2":
print(df.loc["day2"])
```
